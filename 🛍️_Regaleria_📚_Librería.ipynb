Vas a desarrollar un programa en Python que cumpla con las siguientes caracter√≠sticas:
Requerimientos:
Ingreso de datos de productos: El sistema debe permitir ingresar datos b√°sicos de los productos: nombre, categor√≠a, y precio (sin centavos). Estos datos deben almacenarse en una lista, donde cada cliente sea representado/a como una sublista de tres elementos (nombre, categor√≠a, y precio).
Visualizaci√≥n de productos registrados: El programa debe incluir una funcionalidad para mostrar en pantalla todos los productos ingresados. La informaci√≥n debe presentarse de manera ordenada y legible, con cada producto numerado.
B√∫squeda de productos: El sistema debe permitir buscar productos por su nombre. Si encuentra coincidencias, debe mostrar la informaci√≥n completa de los productos que coincidan. Si no hay coincidencias, debe informar que no se encontraron resultados.
Eliminaci√≥n de productos: El sistema debe permitir eliminar un producto de la lista, identific√°ndolo por su posici√≥n (n√∫mero) en la lista.
Requisitos
Usar listas para almacenar y gestionar los datos.
Incorporar bucles while y for seg√∫n corresponda.
Validar entradas del usuario o usuaria, asegur√°ndote de que no se ingresen datos vac√≠os o incorrectos.
Utilizar condicionales para gestionar las opciones del men√∫ y las validaciones necesarias.
Presentar un men√∫ que permita elegir entre las funcionalidades disponibles: agregar productos, visualizar productos, buscar productos y eliminar productos.
El programa debe continuar funcionando hasta que se elija una opci√≥n para salir.
Consejos
Us√° lo aprendido sobre listas y bucles para gestionar los datos y recorrerlos.
Record√° validar las entradas utilizando condicionales.
Utiliz√° las herramientas vistas para organizar y presentar la informaci√≥n de manera clara.
Ejemplo del men√∫ de opciones:
Sistema de Gesti√≥n B√°sica De Productos
1. Agregar producto
2. Mostrar productos
3. Buscar producto
4. Eliminar producto
5. Salir


```python
# Mi lista de productos: [Nombre, Categor√≠a, Precio]
productos = [
    ["Servicio de Copiado - Hoja a color", "Servicios", 500],
    ["Servicio de Copiado - Hoja B/N", "Servicios", 200],
    ["Cuaderno A4", "Papeler√≠a", 5500],
    ["Papel Lunares 2 Hojas Colores Surtidos1", "Papeler√≠a", 1500],
    ["Bloc De Dibujo Nro:5 Blanco 24 Unidades", "Papeler√≠a", 7190],
    ["Lapicera Parker", "Regaler√≠a", 3500],
    ["Bol√≠grafo FILGO Gel Pop Glitter 10 Unidades", "Regaler√≠a", 9500],
    ["Agenda Pocket Galaxia", "Regaler√≠a", 9990],
    ["Cartuchera Objetos", "Regaler√≠a", 8710]
]

def mostrar_menu():
    print("=" * 50)
    print("     üõçÔ∏è Regaler√≠a & üìö Librer√≠a      ")
    print("=" * 50)
    print("1. Agregar producto")
    print("2. Ver productos")
    print("3. Buscar producto")
    print("4. Eliminar producto")
    print("5. Salir")
    print("=" * 50)

opcion = ""
while opcion != 5:
    mostrar_menu()
    opcion = input("Seleccion√° una opci√≥n (1-5): ").strip()

    while not(opcion.isdigit() and 1 <= int(opcion) <= 5):
        print("‚ùå Opci√≥n inv√°lida.")
        opcion = input("Seleccion√° una opci√≥n (1-5): ").strip()
    opcion = int(opcion)

    match opcion:
        case 1:
            print("======= üìù Agregar producto =======")
            nombre = input("Nombre del producto: ").strip().title()
            categoria = input("Categor√≠a (Papeler√≠a, Regaler√≠a, Servicios): ").strip().title()
            precio = input("Precio sin centavos: ").strip()

            datos_ok = True

            if not nombre or not categoria or not precio:
                print("‚ùå Todos los campos son obligatorios.")
                datos_ok = False

            if datos_ok and not precio.isdigit():
                print("‚ùå El precio debe ser un n√∫mero entero.")
                datos_ok = False
            else:
                precio = int(precio)

            if datos_ok:
                producto = [nombre, categoria, precio]
                productos.append(producto)
                print(f"‚úî Producto '{nombre}' agregado.")
            else:
                print("‚ùå No se pudo agregar el producto.")

        case 2:
            print("======= üì¶ Lista de productos =======")
            if not productos:
                print("‚ùå No hay productos registrados.")
            else:
                print("\nüìã {:^60}".format("LISTADO DE PRODUCTOS"))
                print("-" * 60)
                print("{:<3} {:<35} {:<12} {:<8}".format("N¬∞", "Producto", "Categor√≠a", "Precio"))
                print("-" * 60)

                for i, producto in enumerate(productos, 1):
                    print("{:<3} {:<35} {:<12} ${:<8}".format(i, producto[0], producto[1], producto[2]))

        case 3:
            print("======== üîç Buscar producto ============")
            if not productos:
                print("‚ùå No hay productos registrados.")
            else:
                nombre_buscar = input("üîç Ingres√° el nombre del producto: ").strip().title()
                encontrados = [p for p in productos if p[0] == nombre_buscar]

                if encontrados:
                    print("üìù Resultados encontrados:")
                    print("-" * 60)
                    print("{:<3} {:<35} {:<12} {:<8}".format("N¬∞", "Producto", "Categor√≠a", "Precio"))
                    print("-" * 60)
                    for i, producto in enumerate(encontrados, 1):
                        print("{:<3} {:<35} {:<12} ${:<8}".format(i, producto[0], producto[1], producto[2]))
                else:
                    print(f"‚ùå No se encontr√≥ ning√∫n producto llamado '{nombre_buscar}'.")

        case 4:
            print("======== üóëÔ∏è Eliminar producto ============")
            if not productos:
                print("‚ùå No hay productos registrados.")
            else:
                print("\nüìã {:^60}".format("LISTADO DE PRODUCTOS"))
                print("-" * 60)
                print("{:<3} {:<35} {:<12} {:<8}".format("N¬∞", "Producto", "Categor√≠a", "Precio"))
                print("-" * 60)

                for i, producto in enumerate(productos, 1):
                    print("{:<3} {:<35} {:<12} ${:<8}".format(i, producto[0], producto[1], producto[2]))

                pos = input("N√∫mero del producto a eliminar: ").strip()
                while not(pos.isdigit() and 1 <= int(pos) <= len(productos)):
                    print("‚ùå Ingres√° un n√∫mero v√°lido.")
                    pos = input("N√∫mero del producto a eliminar: ").strip()

                eliminado = productos.pop(int(pos) - 1)
                print(f"üóë Producto '{eliminado[0]}' eliminado.")

print("üëã ¬°Gracias por usar el sistema de la librer√≠a!")
```
